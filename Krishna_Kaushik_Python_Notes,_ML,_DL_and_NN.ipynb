```python
# koeyword variable lenght argument.
# keyword variable input argument acts as a dictionary.
def pass_students(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>33:
      pass_student_list.append(name)
  return pass_student_list
```

```python
#transformer: it is a type of neural network to handle sequence to sequence data
#application:- machine translation, question answering, text summarization.
#Attention mechanism enable neural network to select focus on specefic part of input sequence.
#Word Embedding: is a natural language processing (NLP) technique that convert word into numerical vector.
# That allow computer to understand their meaning and relationship of the word.
# These numerical vector are dense form.
# Eg:-   1. TF-IDF
   #     2. Word 2 uec
    #    3. Glove
     #   4. BERT
# Their are two types of word embedding:
  # 1. static word embedding:- provide single, fixed vector, of each word.
                    #           It is simple and fast one.
                    #           Self attension mechnism in contextual word embedding.
  # 2. contextual word embedding:- generate a unique vector for a word based on the specefic sentance.
                    #              they are more accurate because they can capture different meaning of a word such as the word "Bank" in river bank VS money bank.
  # 3. Self attention mechanism :- is build to convert static word embedding into contextual word embedding.
  # Positional encoding:- provides transformer with information about the order of tokens in a sequence.
  # multihead attension mechanism :- is an extention of self attension mechanism used in the tranformer architecture which allows the model to focus on diferent part of the input sequence for multiple purpose.
  # layer normalization:- converts into standard normal distribution. ( mean=0, standard deviation=1)
  # feed forward layer:- thsi layer applied in both part of the transformer encoder as well as decoder. it represent non linearty and capture complex pattern of the neural network.
  # Soft max layer:- convert vectors raw score into probability distribution.
  # their are two roles:-
    # 1. attension mechanism
    # 2. final output classification
  # masked multihead attension mechanism:- is used in the decoder of a transformer to prevent the model from seeing the future when predicting the next token in a sequence.
    # next work prediction
```
